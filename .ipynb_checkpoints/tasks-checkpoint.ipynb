```python
# Определить максимальную длину последовательности подряд идущих 1 в бинарном векторе.


def get_max_seq(mas):
    cur_seq = 0
    max_seq = 0

    for i in mas:
        if i == 1:
            cur_seq += 1
            max_seq = max(cur_seq, max_seq)
        else:
            cur_seq = 0

    return max_seq


print(get_max_seq([]))
print(get_max_seq([0, 0, 0, 0, 0]))
print(get_max_seq([1, 1, 1, 1, 1]))
print(get_max_seq([1, 1, 1, 0, 1, 1, 1, 1]))
```

```python
# Определить являются ли две строка анаграммами (отличаются только порядком следования букв).


from collections import Counter
from collections import defaultdict


def is_anagr(str_0, str_1):
    counter_0 = Counter(str_0)
    counter_1 = Counter(str_1)
    return counter_0 == counter_1


print(is_anagr('qwer', 'rewq'))
print(is_anagr('qwer', 'rewqq'))
print(is_anagr('qwweeeerrr', 'weeeerqrrw'))


def dict_from_str(s):
    res_dict = defaultdict(int)
    for i in s:
        res_dict[i] += 1
    return res_dict

def is_anagr_2(str_0, str_1):
    return dict_from_str(str_0) == dict_from_str(str_1)


print(is_anagr_2('qwer', 'rewq'))
print(is_anagr_2('qwer', 'rewqq'))
print(is_anagr_2('qwweeeerrr', 'weeeerqrrw'))
```

```python
# Сгенерировать все возможные корректные скобочные последовательности для заданного n.


def generate(cur, opened, closed, n):
    if len(cur) == 2 * n:
        print(cur)
        return
    if opened < n:
        generate(cur + '(', opened + 1, closed, n)
    if closed < opened:
        generate(cur + ')', opened, closed + 1, n)

def gen_parens(n):
    generate("", 0, 0, n)


print('for 0:')
gen_parens(0)
print('for 1:')
gen_parens(1)
print('\nfor 2:')
gen_parens(2)
print('\nfor 3:')
gen_parens(3)
```

```python
# Допустима ли скобочная последовательность.


def is_correct(str_in):
    stack = list()
    for i in str_in:
        if stack and stack[len(stack) - 1] == '(' and i == ')':
            stack.pop()
            continue
        stack.append(i)
    if len(stack) == 0:
        print('CORRECT')
    else:
        print('NOT CORRECT')


is_correct('()')
is_correct('())')
is_correct('(()((())()))()')
is_correct('((())')
```

```python
####################################################################################################
# 356. Line Reflection (medium)
####################################################################################################
# Given n points on a 2D plane, find if there is such a line parallel to the y-axis 
# that reflects the given points symmetrically.
#
# In other words, answer whether or not if there exists a line that after reflecting 
# all points over the given line, the original points' set is the same as the reflected ones.
#
# Note that there can be repeated points.
####################################################################################################


from typing import List
import sys


class Solution:
    def isReflected(self, points: List[List[int]]) -> bool:
        max_x = -sys.maxsize - 1
        min_x = sys.maxsize
        points_set = set()

        for p in points:
            max_x = max(max_x, p[0])
            min_x = min(min_x, p[0])
            point = f'{p[0]}, {p[1]}'
            points_set.add(point)

        coord_sum = max_x + min_x

        for p in points:
            point = f'{coord_sum - p[0]}, {p[1]}'
            if point not in points_set:
                return False

        return True


print(Solution().isReflected([[-1, 1], [1, 1]]))
print(Solution().isReflected([[-1, 1], [1, -1]]))
```

```python
####################################################################################################
# 849. Maximize Distance to Closest Person (medium)
####################################################################################################
# You are given an array representing a row of seats where seats[i] = 1 represents a person sitting in the ith seat, 
# and seats[i] = 0 represents that the ith seat is empty (0-indexed).
#
# There is at least one empty seat, and at least one person sitting.
#
# Alex wants to sit in the seat such that the distance between him and the closest person to him is maximized. 
#
# Return that maximum distance to the closest person.
####################################################################################################


class Solution:
    def maxDistToClosest(self, seats: List[int]) -> int:
        first = True
        max_dist = 0
        cur_dist = 0
        for i, val in enumerate(seats):
            if val == 0:
                cur_dist += 1
                if i == len(seats) - 1:
                    max_dist = max(max_dist, cur_dist)
                max_dist = max(max_dist, int((cur_dist + 1) / 2))
                continue
            if val == 1:
                if first:
                    max_dist = cur_dist
                    cur_dist = 0
                    first = False
                if max_dist > int((cur_dist + 1) / 2):
                    cur_dist = 0
                else:
                    max_dist = int((cur_dist + 1) / 2)
                    cur_dist = 0
                continue
        return max_dist


print(Solution().maxDistToClosest([1,0,0,0,1,0,1]))
print(Solution().maxDistToClosest([1,0,0,0]))
print(Solution().maxDistToClosest([0,1]))
```

```python
####################################################################################################
# 228. Summary Ranges (easy)
####################################################################################################
# You are given a sorted unique integer array nums.
#
# A range [a,b] is the set of all integers from a to b (inclusive).
#
# Return the smallest sorted list of ranges that cover all the numbers in the array exactly. 
# That is, each element of nums is covered by exactly one of the ranges, 
# and there is no integer x such that x is in one of the ranges but not in nums.
#
# Each range [a,b] in the list should be output as:
#
# * "a->b" if a != b
# * "a" if a == b
####################################################################################################


class Solution:
    def summaryRanges(self, nums: List[int]) -> List[str]:
        res_list = list()
        if not nums:
            return []
        first = nums[0]
        second = nums[0]
        for i in nums[1:]:
            if i - second == 1:
                second = i
            else:
                if first != second:
                    res_list.append(f'{first}->{second}')
                else:
                    res_list.append(f'{first}')
                first = i
                second = i
        else:
            if first != second:
                res_list.append(f'{first}->{second}')
            else:
                res_list.append(f'{first}')
        return res_list


print(Solution().summaryRanges([0,1,2,4,5,7]))
print(Solution().summaryRanges([0,2,3,4,6,8,9]))
```

```python
####################################################################################################
# 1493. Longest Subarray of 1's After Deleting One Element (medium)
####################################################################################################
# Given a binary array nums, you should delete one element from it.
#
# Return the size of the longest non-empty subarray containing only 1's in the resulting array.
# Return 0 if there is no such subarray.
####################################################################################################


class Solution:
    def longestSubarray(self, nums: List[int]) -> int:
        # Start index of window
        left = 0
        # Count of 0 in current window
        count = 0
        # Max count of 1
        max_count = 0
 
        # Keep window [left..i] with count of 0 <= 1
        for i in range(len(nums)):
            if nums[i] == 0:
                count = count + 1

            if count == 2:
                while nums[left]:
                    left = left + 1

                left = left + 1
                count = 1

            if i - left > max_count:
                max_count = i - left
 
        return max_count


print(Solution().longestSubarray([1,1,0,1]))
print(Solution().longestSubarray([0,1,1,1,0,1,1,0,1]))
```

```python
####################################################################################################
# 161. One Edit Distance (medium)
####################################################################################################
# Given two strings s and t, return true if they are both one edit distance apart, otherwise return false.
#
# A string s is said to be one distance apart from a string t if you can:
#
# - Insert exactly one character into s to get t.
# - Delete exactly one character from s to get t.
# - Replace exactly one character of s with a different character to get t.
####################################################################################################


class Solution:
    def isOneEditDistance(self, s: str, t: str) -> bool:
        if s == t: 
            return False

        i = 0
        while i < min(len(s), len(t)):
            if s[i] == t[i]:
                i += 1
            else:
                break
        return s[i + 1:] == t[i + 1:] or s[i:] == t[i + 1:] or s[i + 1:] == t[i:]


print(Solution().isOneEditDistance('', ''))
print(Solution().isOneEditDistance('ab', 'acb'))
print(Solution().isOneEditDistance('acb', 'ab'))
print(Solution().isOneEditDistance('acb', 'aab'))
```

```python
####################################################################################################
# 443. String Compression (medium)
####################################################################################################
# Given an array of characters chars, compress it using the following algorithm:
#
# Begin with an empty string s. For each group of consecutive repeating characters in chars:
#
# - If the group's length is 1, append the character to s.
# - Otherwise, append the character followed by the group's length.
#
# The compressed string s should not be returned separately, but instead, 
# be stored in the input character array chars. Note that group lengths that are 10 or longer 
# will be split into multiple characters in chars.
#
# After you are done modifying the input array, return the new length of the array.
#
# You must write an algorithm that uses only constant extra space.
####################################################################################################


class Solution:
    def compress(self, chars: List[str]) -> int:
        res_list = list()
        count = 0
        cur_char = chars[0]

        for idx, i in enumerate(chars):
            if i == cur_char:
                count += 1
            if i != cur_char or idx == len(chars) - 1:
                res_list.append(cur_char)
                if count >= 10:
                    for num in str(count):
                        res_list.append(num)
                elif count != 1:
                    res_list.append(str(count))
                if i != cur_char and idx == len(chars) - 1:
                    res_list.append(i)
                cur_char = i
                count = 1

        chars[:] = res_list
        return len(res_list)


l = ["a","a","b","b","c","c","c"]
print(Solution().compress(l), l)
l = ["a"]
print(Solution().compress(l), l)
l = ["a","b","b","b","b","b","b","b","b","b","b","b","b"]
print(Solution().compress(l), l)
l = ["a","a","b","b","c"]
print(Solution().compress(l), l)
```

```python
####################################################################################################
# 3. Longest Substring Without Repeating Characters (medium)
####################################################################################################
# Given a string s, find the length of the longest substring without repeating characters.
####################################################################################################


class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        res = 1 if len(s) else 0
        for i in range(len(s)):
            tmp_str = s[i]
            for j in range(i + 1, len(s)):
                if s[j] not in tmp_str:
                    tmp_str += s[j]
                else:
                    if len(tmp_str) > res:
                        res = len(tmp_str)
                    break
            else:
                if len(tmp_str) > res:
                    res = len(tmp_str)
                    break

        return res


print(Solution().lengthOfLongestSubstring("abcabcbb"))
print(Solution().lengthOfLongestSubstring("bbbbb"))
print(Solution().lengthOfLongestSubstring("pwwkew"))
```

```python
####################################################################################################
# 1. Two Sum (easy)
####################################################################################################
# Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
#
# You may assume that each input would have exactly one solution, and you may not use the same element twice.
#
# You can return the answer in any order.
####################################################################################################


class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i, val_1 in enumerate(nums):
            for j in range(i, len(nums)):
                if i != j:
                    if target == val_1 + nums[j]:
                        return [i, j]


print(Solution().twoSum([2,7,11,15], 9))
print(Solution().twoSum([3,2,4], 6))
print(Solution().twoSum([3,3], 6))
```

```python
####################################################################################################
# 2. Add Two Numbers (medium)
####################################################################################################
# You are given two non-empty linked lists representing two non-negative integers. 
# The digits are stored in reverse order, and each of their nodes contains a single digit. 
# Add the two numbers and return the sum as a linked list.
#
# You may assume the two numbers do not contain any leading zero, except the number 0 itself.
####################################################################################################


# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        num1 = ""
        num2 = ""

        while (l1.next):
            num1 += str(l1.val)
            l1 = l1.next
        else:
            num1 += str(l1.val)

        while (l2.next):
            num2 += str(l2.val)
            l2 = l2.next
        else:
            num2 += str(l2.val)

        res = str(int(num1[::-1]) + int(num2[::-1]))[::-1]

        ln = ListNode()
        tmp = ln
        for n, i in enumerate(res):
            tmp.val = int(i)
            if n + 1 < len(res):
                tmp2 = ListNode()
                tmp.next = tmp2
                tmp = tmp.next

        return ln
```

```python
####################################################################################################
# 21. Merge Two Sorted Lists (easy)
####################################################################################################
# You are given the heads of two sorted linked lists list1 and list2.
#
# Merge the two lists in a one sorted list. The list should be made by splicing together the nodes of the first two lists.
#
# Return the head of the merged linked list.
####################################################################################################


# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        res_list = None

        if list1 or list2:
            res_list = ListNode()

        tmp = res_list

        while list1 and list2:
            if list1.val < list2.val:
                tmp.val = list1.val
                list1 = list1.next
            else:
                tmp.val = list2.val
                list2 = list2.next

            tmp.next = ListNode()
            tmp = tmp.next

        while list1:
            tmp.val = list1.val
            list1 = list1.next
            if list1:
                tmp.next = ListNode()
                tmp = tmp.next

        while list2:
            tmp.val = list2.val
            list2 = list2.next
            if list2:
                tmp.next = ListNode()
                tmp = tmp.next

        return res_list
```

```python
####################################################################################################
# 8. String to Integer (atoi) (medium)
####################################################################################################
# Implement the myAtoi(string s) function, which converts a string to a 32-bit signed integer 
# (similar to C/C++'s atoi function).
#
# The algorithm for myAtoi(string s) is as follows:
#
# Read in and ignore any leading whitespace.
#
# Check if the next character (if not already at the end of the string) is '-' or '+'. 
# Read this character in if it is either. This determines if the final result is negative or positive respectively. 
# Assume the result is positive if neither is present.
#
# Read in next the characters until the next non-digit character or the end of the input is reached. 
# The rest of the string is ignored.
#
# Convert these digits into an integer (i.e. "123" -> 123, "0032" -> 32). If no digits were read, then the integer is 0. 
# Change the sign as necessary (from step 2).
#
# If the integer is out of the 32-bit signed integer range [-2^31, 2^31 - 1], 
# then clamp the integer so that it remains in the range. Specifically, 
# integers less than -2^31 should be clamped to -2^31, and integers greater than 2^31 - 1 should be clamped to 2^31 - 1.
# Return the integer as the final result.
#
# Note:
#
# - Only the space character ' ' is considered a whitespace character.
# - Do not ignore any characters other than the leading whitespace or the rest of the string after the digits.
####################################################################################################


class Solution:
    def myAtoi(self, s: str) -> int:
        s = s.strip()
        negative = False
        res = ''
        idx = 0

        if not s:
            return 0

        if s[0] == '-':
            negative = True
            idx = 1
        elif s[0] == '+':
            idx = 1

        for i in s[idx:]:
            if not i.isdigit():
                break
            res += i

        if not res:
            return 0

        res_int = int(res)
        res_int = -res_int if negative else res_int

        if res_int < -2**31:
            return -2**31

        if res_int > 2**31 - 1:
            return 2**31 - 1

        return res_int


print(Solution().myAtoi("42"))
print(Solution().myAtoi("   -42"))
print(Solution().myAtoi("4193 with words"))
```

```python
####################################################################################################
# 7. Reverse Integer (medium)
####################################################################################################
# Given a signed 32-bit integer x, return x with its digits reversed. 
# If reversing x causes the value to go outside the signed 32-bit integer range [-231, 231 - 1], then return 0.
#
# Assume the environment does not allow you to store 64-bit integers (signed or unsigned).
####################################################################################################


class Solution:
    def reverse(self, x: int) -> int:
        is_positive = True if x > 0 else False
        x = abs(x)
        res = 0

        while x > 0:
            res = res * 10 + x % 10
            x = int(x / 10)

        res = res if is_positive else -res

        if -2**31 <= res <= 2**31 - 1:
            return res

        return 0


print(Solution().reverse(123))
print(Solution().reverse(-123))
print(Solution().reverse(120))
```

```python
####################################################################################################
# 88. Merge Sorted Array (easy)
####################################################################################################
# You are given two integer arrays nums1 and nums2, sorted in non-decreasing order, 
# and two integers m and n, representing the number of elements in nums1 and nums2 respectively.
#
# Merge nums1 and nums2 into a single array sorted in non-decreasing order.
#
# The final sorted array should not be returned by the function, but instead be stored inside the array nums1. 
# To accommodate this, nums1 has a length of m + n, where the first m elements denote the elements that should be merged, 
# and the last n elements are set to 0 and should be ignored. nums2 has a length of n.
####################################################################################################


class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        tmp = nums1[:m]
        tmp.extend(nums2)
        res = sorted(tmp)
        for i in range(m + n):
            nums1[i] = res[i]


nums1 = [1,2,3,0,0,0]
Solution().merge(nums1, 3, [2,5,6], 3)
print(nums1)
nums1 = [1]
Solution().merge(nums1, 1, [], 0)
print(nums1)
nums1 = [0]
Solution().merge(nums1, 0, [1], 1)
print(nums1)
```

```python
# Является ли число степенью двойки.


def is_power_of_two(num: int) -> bool:
    # Такое число имеет только один выставленный бит в своём представлении
    if num & (num - 1):
        return False
    return True


print(is_power_of_two(8))
print(is_power_of_two(2048))
print(is_power_of_two(14))
print(is_power_of_two(5))
```

```python
# Переставить попарно биты в числе.


def bin_to_int(num: str) -> int:
    return int(num, 2)

def int_to_bin(num: int) -> str:
    return bin(num).replace('0b', '')

def swap_bits(num: int) -> str:
    mask_10101010 = bin_to_int('10101010')
    mask_01010101 = bin_to_int('01010101')

    num_1 = num & mask_10101010
    num_2 = num & mask_01010101

    num_1 >>= 1
    num_2 <<= 1

    res = num_1 | num_2

    return int_to_bin(res)


print(swap_bits(bin_to_int('01011100')))
print(swap_bits(bin_to_int('01011001')))
```
